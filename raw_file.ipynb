```python
# machine learning
machine learning is making the computer learn from styding data and statistics.
machine learning is a step into the direction of artificial intelligence (AI).
machine learning is a programme that analyzes data and learns to predict the outcome.

imagine writing a programme to identify cats, dogs, rabbit and goat.
using traditional programming technique, you may need to write rules with a lot of codes to look for edges, curves, patterns, colors and others.

what if in the future you are given another animal to identify and it is a black and white image, you are going to re-write all your codes again.
with machine learning it is made easy such as
we build a model or an engine that gives it lots and lots of data could be tens of thousands, of images (maybe cow, cat, dog, goat etc.)

our model would then find and learn patterns in the input data.
if we. give a new picture of cat or a dog or a horse, its going to tell us with certain level of accuracy.
the more data we input, the more accuracy it would give us.



# machine learning has other applications
we use it in self driven cars
we use it in robotics
we use it in language processing
we use it in vision processing
we use it for forcasting stock market/forest trading
we use it for forcasting weather

# steps involved in machine learning
machine learning project involves the following steps:
1. import the data  which is usually in a CSV format
2. clean the data (remove bugs: duplicates or incomplete files)if theym are alphabetes we need to convert them to numerical values else, we will get wrong result.
3. split: splits the data into two segment (1. testing 2. training) you can reserve 80% for traing, 20% for testing.
4. create our model
** this involves selecting an algorithm to analyze our data
** there are differnt types of algorithm such as, the decision tree, neuro network etc.
** each agorithm has its props and cons in terms of accuracy and performance so the agorithm you choose depends on the problem you are trying to solve in your input data.
** there are libraries you can use that provide this algorithm eg. sciki_learn.

5. train the model
** we feed our training data to our model, it will look for patterns in the data

6. make predictions
** we would ask our model to make predictions(it may not be accurate as a beginer)

7. evaluate and improve
** we evaluate the prediction and make sure they are accurate then we need to go back to our model to either select another algorithm that is going to produce a better result for the kind of problem we are trying to solve 
or we finetune the parameters of our problem
** each algorithm has parameters we can modify to optimize accuracy.




# libriaries in machine learning/ data science
1. numpy-numpy provides multidimentional arrays
2. the pandas library- its a data analysis library thatm provides a concept called data frame
** a data frame is a two dimentional structure similar to an excel spreed sheet:we have rows and coloms
** we can select data in rows and coloms or range or rows and coloms
3. the mathplot lib. _ it is a two dimentionl plotting library for creatin graphs and plots
4. the sicki-learn - wchich is one of the most popular machine learning library that provides common algorithm such as: decision tree, neuro network etc.

import pandas as pd 
sales = pd.read_csv('vgsales.csv')
sales.shape

sales.describe

sales.values

sales.dtypes

sales.head

```
