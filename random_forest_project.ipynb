```python
import streamlit as st
import pandas as pd
# from sklearn.model_selectionmodel import train_test_split
from sklearn.model_selection import train_test_split
# from sklearn.ensemble import RandomForestRegressorn
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.datasets import load_diabetes, load_boston
```

```python
st.set_page_config(page_title='The Machine Learnin App', layout='wide')
```

```python
def build_model(df):
    X = df.iloc[:,:-1] #using all cplume except the las one
    Y = df.iloc[:,-1] #selecting the last column as Y
    
    st.markdown('**1.2. Data Splits**')
    st.write('Training set')
    st.info('X.shape')
    st.info('Test set')
    st.info(Y.shape)
    
    
    st.markdown('**1.3. Variable details**')
    st.write('X variable')
    st.info(list(X.columns))
    st.write('Y variable')
    st.info(Y.name)
    
    #data spliting
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=split_size)
    
    rf = RandomForestClassifier(n_estimators=parameter_n_estimators,
         random_state=parameter_random_state,
         max_features=parameter_max_features,
         criterion=parameter_criterion,
         min_samples_split=parameter_min_samples_split,
         min_samples_leaf=parameter_min_samples_leaf,
         bootstrap=parameter_bootstrap,
         oob_score=parameter_oob_score,
         n_jobs=parameter_n_jobs)
    rf.fit(X_train, Y_train)
    
    
    st.subheader('2. Model Performance')
    
    st.markdown('**2.1. Training set')
    Y_pred_train = rf.predict(X_train)
    st.write('Coefficient of determination ($R^2$):')
    st.info( r2_score(Y_train, Y_pred_train) )
    
    st.write('Mean squared error (MSE):')
    st.info( mean_squared_error(Y_train, Y_pred_train) )
    
    st.markdown('**2.2. Test set**')
    Y_pred_test = rf.predict(X_test)
    st.write('Coefficient of determination ($R^2$):')
    st.info( r2_score(Y_test, Y_pred_test) )
    
    st.subheader('3. Model Parameters')
    st.write(rf.get_params())
    
    
    #----------------#
    st.write("""
    #the machine learning app
    in this implementation, the randomforestclassifier() * Function is used in this
    
    Try adjusting the parameters
    """)
    #side bar to collect user input data-set
    
    with st.sidebar.header('1. Upload your CSV data'):
        uploaded_file = st.sidebar.file_uploader("upload your input CSV file")
        st.sidebar.markdown("""
        [Example CSV input file](https://raw.github.com/dataprofessor/data.git)
        """)
        
    #sidebar specify parameter
    with st.sidebar.subheader('2. Set Parameters'):
        slit_size = st.sidebar.slider('Data slpit ratio(% for Training Set)', 10, 90, 80, 5)
        
    with st.sidebar.subheader('2.1. Learning Parameters'):
        parameter_n_estimators = st.sidebar.slider('Number of estimators(n_estimators)', 0, 1000, 100, 100)
        parameter_max_features = st.sidebar.select_slider('Max feature(max_features)', options=['auto', 'sqrt', 'log2'])
        parameter_min_samples_split = st.sidebar.slider('Minimum numbers of sample required to split an internal node(min_sample_leaf)')
        parameter_min_samples_leaf = st.sidebar.slider('Minimum numbers of sample required to be at a leaf node(Min sample_leaf)')
        
        
    with st.sidebar.subheader('2.2. General Parameter'):
        parameter_random_state = st.sidebar.slider('Seed Number(random_state)', 0, 1000, 42, 1)
        parameter_criterion = st.sidebar.select_slider('Performance measure(criterion)', options=['mse', 'mae']')
        parameter_bootstrap = st.sidebar.select_slider('Bootstrap Samples when building trees(bootsrap)', options=[True, False])
        parameter_oob_score = st.sidebar.select_slider('Whether to use out of bag samples to estimate the R^2 on unseen data(oob_score)')
        parameter_n_jobs = st.sidebar.select_slider('Number of jobs to run in parallel(n_jobs)', options=[1, -1])


#--------------------------------#
#main panel

# dispaly the dataset
st.subheader('1. Dataset')

if uploaded_file is not None:
    df = pd.read_csv(uploaded_file)
    st.markdown('**1.1. Glimpse of dataset**')
    st.write(df)
    build_model(df)
else:
    st.info('Awaiting for CSV file to be uploaded.')
    if st.button('Press to use Dataset Example'):
        #diabities dataset
        #diabetes = load_diabetes()
        #X = pd.DataFrame(diabetes.data, column=diabetes.feat)
        #Y = pd.Series(diabetes.target, name='response')
        #df = pd.concat( [X,Y], axis=1 )
        
        #st.markdown('the diabetes set is used as an example')
        #st.write(df.head(5))
        
        #boston housing data
        boston = load_boston()
        X = pd.DataFrame(boston.data, columns=boston.feature)
        Y = pd.Series(boston.target, name='response')
        df = pd.concat( [X,Y], axis=1 )
                                                       
        
        st.markdown('The Boston housing dataset is used as the ')
        st.write(df.head(5))
        
        build_model(df)
```
